## I) B)

### We extract information from the 'characteristics' column

```python
data['ram'] = data['caracteristics'][data['caracteristics'].str.contains('RAM :')].str.split("RAM : ", -1).str.get(1).str.split("Go", -1).str.get(0).str.replace(' ','')
data['processeur'] = data['caracteristics'][data['caracteristics'].str.contains('Processeur :')].str.split("Processeur : ", 1).str.get(1).str.split('\n', 1).str.get(0)
data['resolution'] = data['caracteristics'][data['caracteristics'].str.contains('Résolution :')].str.split("Résolution : ", 1).str.get(1).str.split('\n', 1).str.get(0)
data['stockage'] = data['caracteristics'][data['caracteristics'].str.contains('Stockage principal :')].str.split("Stockage principal : ", 1).str.get(1).str.split(' Go|\n', 1).str.get(0)
data['processeur graphique'] = data['caracteristics'][data['caracteristics'].str.contains('Processeur graphique : ')].str.split("Processeur graphique : ", 1).str.get(1).str.split('\n', 1).str.get(0)
data['duree fonctionnement'] = data['caracteristics'][data['caracteristics'].str.contains('Durée de fonctionnement : ')].str.split("Durée de fonctionnement : ", 1).str.get(1).str.split('\n', 1).str.get(0)
```

### We associate a brand for each article

```python
data = data.reset_index(drop = True)

titles = data['title']
brands=[]
index=[]

brand_list = ['microsoft','htf', 'ofeli','panasonic', 'hsw','samsung', 'apple',
              'hp','compaq','dell','bell','acer','emachines','asustoshiba','samsung'
              ,'sony','lenovo','asus','razer']

for i,title in enumerate(titles):
    spl = title.split(' ')
    for word in spl:
        if word.lower() in brand_list:
            brands.append(word.lower())
            index.append(i)

df_brands = [None for k in range(len(titles))]
c=0
for k in index:
    df_brands[k]=brands[c]
    c+=1

data['brands'] = pd.DataFrame(df_brands)
data
```

```python
dells = data[data['brands']=='dell'].reset_index(inplace=False)
dells['price']= dells['price'].str.replace(',','.').astype(float)
dells = dells[dells['price']>50].reset_index(inplace=False)
dells['title']=dells['title'].str.lower()
dell_liste = ['xspiron', 'xps', 'serie g', 'alienware', 'latitude']

models=[]
index = []
for i,ligne in enumerate(dells['title']):
    for k in range(len(dell_liste)):
        if dell_liste[k] in ligne:
            models.append(dell_liste[k])
            index.append(i)
dells['modele']= pd.DataFrame(models,index)
```

## II) Scheme of price dispersion

### A) For DELLS

#### We group computers by model, each element of comparison must be identical

```python
dells_gb = dells.groupby(['modele','ram','resolution','stockage','processeur graphique'])
```

```python
articles = []
for i,x in enumerate(dells_gb):
    y=dells_gb.get_group(x[0])
    articles.append(y)

c,m=0,0
group, not_group = 0,0

for i in range(len(articles)):
    if len(articles[i]['price'])<5: continue  # We wan't enough data to avoid spurrious correlation
    c+=1
    maxi,mini = max(articles[i]['price']),min(articles[i]['price'])
    delta = maxi-mini
    m+=delta
    for p in articles[i]['price']:
        if p<mini+0.2*delta or p>maxi-0.2*delta:
            group+=1
        else: not_group+=1
            
plt.pie([group,not_group]
        ,labels=['Computers around min and max \n'+str(int(100*group/(group+not_group)))
                 ,'Computers far from min and max \n'+str(100-int(100*group/(group+not_group)))]
        ,shadow = True,startangle=90.0,radius=2.5)

print("     Proportion of computers (in average) which are around the extremum \n                    value (=20% above or below)")
plt.show()  

if c!=0:
    m_dell=m/c 
articles_dell = articles
```

```python
print("L'écart moyen entre le prix maximum et minimum pour un ordinateur dell est de",m_dell,"€")
```

We can clearly see that there is a scheme that is profiling, most computer's price are either condensed around the low prices or the low prices but there are not a lot of computer of which the price is between those areas.


If the distribution was uniform, the orange part of the pie would be of 60% (=100-2*20) but it represents just a bit more than 20%

We can interpret that people devalue the price of the good as soon at it is not brand new. Therefore, just a little bit of use of a little flaw that shows that the computer isn't new can lead to a great loss while selling.

## III) Correlation between price and price dispersion

### A) Does it exist?

```python
delta_mac = []
delta_dell = []
av_price_mac = []
av_price_dell = []

for i in range(len(articles_mac)):
    if len(articles_mac[i].sort_values('price')['price'])<20: continue
    sorted_mac = articles_mac[i].sort_values('price')
    delta_mac.append(max(sorted_mac['price'])-min(sorted_mac['price']))
    av_price_mac.append(np.mean(sorted_mac['price']))
    
for i in range(len(articles_dell)):
    if len(articles_dell[i].sort_values('price')['price'])<5: continue
    sorted_dell = articles_dell[i].sort_values('price')
    delta_dell.append(max(sorted_dell['price'])-min(sorted_dell['price']))
    av_price_dell.append(np.mean(sorted_dell['price']))
    
delt_mac, ave_mac = [],[]
delt_dell, ave_dell = [],[]
for i in range(10,len(delta_mac)-10): 
    delt_mac.append(np.mean(delta_mac[i-10:i+10]))
    ave_mac.append(np.mean(av_price_mac[i-10:i+10]))
for i in range(3,len(delta_dell)-3): 
    delt_dell.append(np.mean(delta_dell[i-3:i+3]))
    ave_dell.append(np.mean(av_price_dell[i-10:i+10]))
    
abs_mac = pd.DataFrame(np.transpose(np.asarray([delt_mac,ave_mac])),columns=['delt_mac','ave_mac'])
abs_mac = abs_mac.sort_values(['ave_mac'])
abs_mac = abs_mac.rolling(5,axis=0).sum()/5

abs_dell = pd.DataFrame(np.transpose(np.asarray([delt_dell,ave_dell])),columns=['delt_dell','ave_dell'])
abs_dell = abs_dell.sort_values(['ave_dell'])
abs_dell = abs_dell.rolling(5,axis=0).sum()/5


plt.figure(1,figsize=(20,7))
plt.xlabel('Computers by growing prices')
plt.subplot(1,2,1)
plt.grid()
plt.xlabel('Average price of the model (€)')
plt.ylabel('maximum price dispersion (€)')
plt.title('Mac')
plt.plot(abs_mac['ave_mac'],abs_mac['delt_mac'])
plt.subplot(1,2,2)
plt.xlabel('Average price of the model (€)')
plt.ylabel('maximum price dispersion (€)')
plt.title('Dell')
plt.grid()
plt.plot(abs_dell['ave_dell'],abs_dell['delt_dell'])

print('                     Evolution of price dispersion with the average price of the good')
plt.show()
```

We can clearly see that the price dispersion seems to diminuish with the average price of the good. The plot is less smooth for macs because there are more data and therefore more randomness in somes values.

Wan can interpret that sellers give more care when the price is high. It is indeed understandable, for a medium computer you don't make the effort of finding the exact value but for a high tech computer (and therefore a computer which might have more visibility) it is logical to do more research.

Moreover, for cheap computers it might be complicated to find the valuation because of how it has been used, of the age the computer might have etc...

### A) Can we find a model for it?

```python
abs_dell=abs_dell.dropna()
abs_dell['delt_dell']
```

```python
from sklearn.linear_model import LinearRegression as lr
reg_dell = lr().fit(np.asarray(abs_dell['ave_dell']).reshape(-1, 1)
               ,np.asarray(list(abs_dell['delt_dell'])))
reg_mac = lr().fit(np.asarray(abs_mac['ave_mac'][5:-7]).reshape(-1, 1)
               ,np.asarray(list(abs_mac['delt_mac'][5:-7])))
```

```python
reg_y_dell = reg_dell.intercept_ + reg_dell.coef_[0]*abs_dell['ave_dell']
reg_y_mac = reg_mac.intercept_ + reg_mac.coef_[0]*abs_mac['ave_mac']


plt.figure(1,figsize=(20,7))

plt.subplot(1,2,1)
plt.grid()
plt.xlabel('Average price of the model (€)')
plt.ylabel('maximum price dispersion (€)')
plt.title('Mac')
plt.plot(abs_mac['ave_mac'],abs_mac['delt_mac'])
plt.plot(abs_mac['ave_mac'],reg_y_mac)

plt.subplot(1,2,2)
plt.xlabel('Average price of the model (€)')
plt.ylabel('maximum price dispersion (€)')
plt.title('Dell')
plt.grid()
plt.plot(abs_dell['ave_dell'],abs_dell['delt_dell'])
plt.plot(abs_dell['ave_dell'],reg_y_dell)
plt.show()

print(reg_dell.score((np.asarray(abs_dell['ave_dell']).reshape(-1, 1)
                     ),np.asarray(list(abs_dell['delt_dell']))))

print("FOR THE MAC REGRESSION")
print("The coeffecient of the mac regression is: ",reg_mac.coef_[0])
print("The intercept of the mac regression is: ",reg_mac.intercept_,)
print("The number of data element is:", len(abs_mac['ave_mac']),'\n')
print("FOR THE DELL REGRESSION")
print('The determiation coeficient for this regression is R^2:',reg_dell.score((np.asarray(abs_dell['ave_dell']).reshape(-1, 1)),np.asarray(list(abs_dell['delt_dell']))))
print("The coeffecient of the dell regression is: ",reg_dell.coef_[0])
print("The intercept of the dell regression is: ",reg_dell.intercept_)
print("The number of data element is:", len(abs_dell['ave_dell']),'. It is not high because of the fact that there are many models for dell, and therefore only a few data for each model. Since we filter models with less than 5 articles to avoid errors, there arent many data element left.')
```

This regression validates our theory since the slope of each curve is negative.
However, this method can be criticized because of the lack of data we have for the dell regression. Despite some additional efforts, we didn't manage to fix this issue.

### a) Preprocessing

```python
def show(x):
    display(HTML(x.to_html()))

macs_air = macs.groupby(['modele','ram','stockage','seller'],as_index=False).mean().iloc[19:42]
macs_pro = macs.groupby(['modele','ram','stockage','seller'],as_index=False).mean().iloc[49:84]
show(macs.groupby(['modele','ram','stockage','seller']).mean())
```

### For Macbook pros

```python
macs_pro.head()
```

```python
sellers = macs_pro['seller'].unique()
dico = {'AdopteUnMac le meilleur du Mac reconditionne':0,
       'E Recycle':0,
        'ELEKTROMANIA':0,
        'Mackfan':0,
        'Okamac':0,
        'thescorpionkings':0,
        'E-nova':0,
        'EBT ELECTRONICS':0,
        'Hobbystore':0,
        'PC4U':0,
        'VDD - Vente du Diable':0}
occur = {'AdopteUnMac le meilleur du Mac reconditionne':0,
       'E Recycle':0,
        'ELEKTROMANIA':0,
        'Mackfan':0,
        'Okamac':0,
        'thescorpionkings':0,
        'E-nova':0,
        'EBT ELECTRONICS':0,
        'Hobbystore':0,
        'PC4U':0,
        'VDD - Vente du Diable':0}

L = np.transpose(np.vstack([macs_pro['seller'],macs_pro['price']]))



res = [L[0:10],L[12:19],L[19:]]

results =[]
for i in range(3):
    df = pd.DataFrame()
    df['seller']=res[i][:,0]
    df['price']=res[i][:,1]
    results.append(df)
    


for r in results: 
    r = r.sort_values('price')
    for k in range(len(r)):
        dico[list(r['seller'])[k]]+=len(r)-k
        occur[list(r['seller'])[k]]+=1
for c in dico:
    dico[c]/=occur[c]

    

data = list(dico.items())
res = pd.DataFrame(np.asarray(data),columns=['seller','expensiveness'])
res['expensiveness'] = pd.to_numeric(res['expensiveness'])
res = res.sort_values('expensiveness')
res
```

### b) Analysis

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

etiquettes = res['seller']
valeurs = res['expensiveness']

# Affichage des données
ax.bar(etiquettes, valeurs, color=[
       'red', 'blue', 'orange', 'green', 'gray'], edgecolor="black")

plt.title("Expensiveness of each buyer:")  
plt.ylabel('Average ranking by decreasing price') 
plt.xlabel('Sellers')
plt.xticks(rotation=75)
plt.show()
```

What we call 'expensiveness' here is the average ranking of the buyer for different models of computers
We can clearly observe that the expensiveness can be very different from one to another. Therefore some sellers who are selling the same computer can fix different prices and this is the origin of price dispersion.

We can interprete that some sellers have a very good reputation (either by ratings or because it is famous). 

Another theory would be that there is a circle effect: the more they have a good reputation the more they can sell and therefore the more it is reassuring for buyer. It allows them to gain in reputation and to raise prices as a consequence.

### b) Model

##### The two previous characteristics are interesting, but they seem to be complementary. Indeed, for a given computer model, it is the RAM memory and the storage that make the difference between the models. This is why we will try to model the price of a model and its specificities using these two characteristics.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15,9))


my_cmap = plt.get_cmap('turbo')
ax = fig.add_subplot(projection='3d')
sctt = ax.scatter(macs_pro['ram'], macs_pro['stockage'], macs_pro['price']
                  ,alpha = 0.8,c = (macs_pro['ram']+ macs_pro['stockage']+ macs_pro['price']),cmap = my_cmap)

ax.set_xlabel('RAM')
ax.set_ylabel('Computer storage')
ax.set_zlabel('Price')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)


model = LinearRegression()
model.fit(macs_pro[['ram','stockage']], macs_pro['price'])
y_pred = model.predict(macs_pro[['ram','stockage']])

coefs = model.coef_
intercept = model.intercept_
xs,ys = np.meshgrid(range(0,20), range(700))
zs = xs*coefs[0]+ys*coefs[1]+intercept
ax.plot_surface(xs,ys,zs, alpha=0.5)



#IMPORTANT:
#You can modify the value "angle" above to see the figure from another point of vue
angle = 75
high = 20


ax.view_init(high, angle)
```

Nous pouvons en effet observer une corrélation entre les points qui sont aux alentours du plan de regression.

```python
print('The coefficients of this regression are:',model.coef_[0],' (for ram) and',model.coef_[0],' (for storage)')
print('The intercept of this regression are:',model.intercept_)
print('The determiation coeficient for this regression is:',model.score(macs_pro[['ram','stockage']], macs_pro['price']))
print('the numer of point for this correlation is:',len(macs_pro['ram']))
```

So we can see that these two elements are relevant when combined when determining the price of computers!
